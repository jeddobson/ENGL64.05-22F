```python
from gensim.models import FastText
from nltk.tokenize import sent_tokenize, word_tokenize
import glob
```

```python
class load_sentences:
    def __iter__(self):
        for document in glob.glob("../data/na*/data/texts/*.txt"):
                with open(document, 'r', encoding='utf-8') as fp:
                    text = fp.read()
                    text = text.replace('—',' ')
                    text = text.replace('“',' ')
                    text = text.replace('”',' ')
                    text = text.replace("‘",' ')
                    sentences = sent_tokenize(text)
                    for sent in sentences:
                        tokens = word_tokenize(sent)
                        tokens = [w.lower() for w in tokens]
                        yield tokens
```

```python
model = FastText(vector_size=300, 
                 window=10, 
                 min_count=5,
                 sg=1)
model.build_vocab(corpus_iterable=load_sentences())
total_examples = model.corpus_count
model.train(corpus_iterable=load_sentences(), 
            total_examples=total_examples, 
            epochs=5)
```

```python
model.wv.most_similar(['freedom'],topn=200)
```
